```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
file_path = 'Financial Analytics data.csv'  # Replace with your file path
data = pd.read_csv(file_path)
```

```python
data.head(5)
```

```python
data_cleaned = data.drop(columns=['S.No.', 'Unnamed: 4'])
```

```python
data_cleaned.columns = ['Name', 'Market_Cap_Crore', 'Quarterly_Sales_Crore']
```

```python
data_cleaned = data_cleaned.dropna(subset=['Market_Cap_Crore'])
```

```python
median_sales = data_cleaned['Quarterly_Sales_Crore'].median()
data_cleaned['Quarterly_Sales_Crore'].fillna(median_sales, inplace=True)
```

```python
top_10_market_cap = data_cleaned.sort_values(by='Market_Cap_Crore', ascending=False).head(10)
print("Top 10 Companies by Market Capitalization:")
print(top_10_market_cap)
```

```python
top_10_sales = data_cleaned.sort_values(by='Quarterly_Sales_Crore', ascending=False).head(10)
print("Top 10 Companies by Quarterly Sales:")
print(top_10_sales)
```

```python
correlation = data_cleaned['Market_Cap_Crore'].corr(data_cleaned['Quarterly_Sales_Crore'])
print(f'Correlation between Market Capitalization and Quarterly Sales: {correlation}')
```

```python
plt.figure(figsize=(10, 6))
data_cleaned['Market_Cap_Crore'].hist(bins=50)
plt.title('Distribution of Market Capitalization')
plt.xlabel('Market Capitalization (Crore)')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(data_cleaned['Market_Cap_Crore'], data_cleaned['Quarterly_Sales_Crore'])
plt.title('Market Capitalization vs Quarterly Sales')
plt.xlabel('Market Capitalization (Crore)')
plt.ylabel('Quarterly Sales (Crore)')
plt.show()
```
